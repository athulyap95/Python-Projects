```python
from bs4 import BeautifulSoup
import requests
```

```python

headers = {
    'User-Agent': 'MyBot/1.0 (athulyavijayan351@gmail.com)'
}
response = requests.get('https://en.wikipedia.org/wiki/The_World%27s_Billionaires', headers=headers)
```

```python
soup = BeautifulSoup(response.text,'html.parser')
soup
```

```python
soup.find('table')
```

```python
soup.find_all('table')
```

```python
table_2025 = soup.find_all('table')[2]
```

```python
table_title = table_2025.find_all('th')
table_title
```

```python
for title in table_title:  
    print(title)
    print('-------------')
    print(title.text.strip())
    print('.............')
```

```python
heading = [title.text.strip() for title in table_title ]
heading
```

```python
import pandas as pd
```

```python
df_2025 = pd.DataFrame(columns = heading)
```

```python
df_2025
```

```python
table_2025
```

```python
column_data = table_2025.find_all('tr')
column_data
```

```python
for row in column_data[1:]:
    rowdata = row.find_all('td')
    data_table = [data.text.strip() for data in rowdata]
    df_2025.loc[df_2025.shape[0]] = data_table
```

```python
df_2025
```

```python
df_2025.to_csv(r'The_world_billionaires.csv',index = False)
```
